```python
import pandas as pd
import os

print("--- 1.Загрузка данных ---")

ratings = pd.read_csv('ratings.csv')
print("Данные с рейтингами (ratings)")
print(ratings.head())
print("\n")

movies = pd.read_csv('movies.csv')
print("Данные с фильмами (movies):")
print(movies.head())
print("\n")

keywords = pd.read_csv('keywords.csv')
print("Данные с ключевыми словами (keywords)")
print(keywords.head())
print("-" * 30 + "\n")


print("--- 2.Классифицируем фильмов по рейтингу ---")

def classify_rating(rating):
    """
    Классифицирует по условиям ДЗ:
    - rating <= 2.0  -> 'низкий рейтинг'
    - rating <= 4.0  -> 'средний рейтинг'
    - rating >= 4.5  -> 'высокий рейтинг'
    """
    if rating <= 2.0:
        return 'низкий рейтинг'
    elif rating <= 4.0:
        return 'средний рейтинг'
    elif rating >= 4.5:
        return 'высокий рейтинг'
    else:
        return 'неизвестный рейтинг'

print("Функция для классификации работает.")


ratings['class'] = ratings['rating'].apply(classify_rating)
print("Столбец 'class' успешно создан.")
print("\nНовая Таблица с рейтингами (первые 10 строк):")
print(ratings.head(10)) # Показываем больше строк для демонстрации разных классов
print("\n")


print("Распределение оценок по категориям:")
print(ratings['class'].value_counts())
print("-" * 30 + "\n")


print("--- 3.Объединение таблиц ratings и movies ---")

full_ratings_info = pd.merge(ratings, movies, on='movieId', how='left')

print("Таблицы успешно объединены. Итого: вид данных (первые 10 строк):")
print(full_ratings_info[['userId', 'title', 'rating', 'class', 'genres']].head(10))
print("-" * 30 + "\n")

print("--- 4.Анализ ключевых слов ---")

print("Топ-5 самых популярных поисковых запросов из keywords.csv:")
print(keywords.head())


mail_keywords = keywords[keywords['keyword'].str.contains('почта|mail', case=False, na=False)]
print("\nПоисковые запросы, содержащие 'почта' или 'mail' (без учета регистра):")
print(mail_keywords)
print("-" * 30 + "\n")

print("Анализ завершен.")
```

```python
import pandas as pd

keywords = pd.read_csv('keywords.csv')

print("---Исходные данные ---")
print(keywords)
print("\n" + "="*40 + "\n")

print("--- 1.Создан гео-классификатор ---")
geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
def classify_by_geo(keyword):
    """
    Проверяет, содержит ли строка 'keyword' название города из geo_data.
    Возвращает название региона или 'undefined'.
    """
    keyword_lower = str(keyword).lower()
    for region, cities in geo_data.items():
        for city in cities:
            if city in keyword_lower:
                return region
    return 'undefined'
    
keywords['region'] = keywords['keyword'].apply(classify_by_geo)

print("Завершено.\n")


print("--- 2.Результат ---")

print("Строки с регионом:")
classified_rows = keywords[keywords['region'] != 'undefined']
print(classified_rows)
print("\n")

print("Таблица с 'region':")
print(keywords)
print("\n")

print("Распределение запросов по регионам:")
print(keywords['region'].value_counts())
```

```python
import pandas as pd

print("--- 1.Загрузка данных ---")
try:
    movies = pd.read_csv('movies.csv')
    ratings = pd.read_csv('ratings.csv')
    print("Файлы movies.csv и ratings.csv успешно загружены.")
    print(f"Загружено {len(movies)} фильмов и {len(ratings)} оценок.")
except FileNotFoundError:
    print("Ошибка: Убедитесь, что файлы movies.csv и ratings.csv находятся в той же папке, что и ваш скрипт.")
    exit()

print("\n" + "="*40 + "\n")

print("--- 2.Определение года выпуска фильмов ---")

years = list(range(1950, 2011))
def production_year(title):
    for year in years:
        if str(year) in title:
            return year
    return 1900

movies['year'] = movies['title'].apply(production_year)

print("Столбец 'year'создан.")

print("\nПримеры фильмов с определенным годом:")
print(movies[movies['year'] != 1900].head())
print("\nПримеры фильмов, для которых год не определился (установлен 1900):")
print(movies[movies['year'] == 1900].head())

print("\n" + "="*40 + "\n")

print("--- 3.Расчет и сортировка среднего рейтинга по годам ---")

merged_data = pd.merge(movies, ratings, on='movieId')
average_rating_by_year = merged_data.groupby('year')['rating'].mean().sort_values(ascending=False)
print("Средний рейтинг фильмов для каждого года (сортировка по убыванию):\n")
print(average_rating_by_year)
```
